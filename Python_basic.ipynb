```python
# 5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each
# Creating a list (lists are mutable)
my_list = [10, 20, 30, 40]

# Creating a tuple (tuples are immutable)
my_tuple = (10, 20, 30, 40)

# Changing an element in the list
# This works because lists are mutable
my_list[1] = 200
print("Modified list:", my_list)

# Trying to change an element in the tuple
# This will cause an error because tuples are immutable
try:
    my_tuple[1] = 200   # Not allowed
except TypeError:
    print("Error: Cannot modify elements of a tuple because it is immutable.")

# Display original tuple to show it remains unchanged
print("Original tuple:", my_tuple)

```

```python
# 6.  Write a function to demonstrate the behavior of mutable and immutable arguments.
# Function that tries to modify both an immutable and a mutable argument
def modify_values(num, lst):
    # num is immutable (int)
    num = num + 10       # This creates a NEW integer object
    # lst is mutable (list)
    lst.append(100)      # This modifies the ORIGINAL list
    return num, lst


# Immutable variable (integer)
x = 5

# Mutable variable (list)
my_list = [1, 2, 3]

# Call function
result_num, result_list = modify_values(x, my_list)

# Printing results
print("After function call:")
print("x (immutable int):", x)               # Will remain unchanged
print("my_list (mutable list):", my_list)    # Will be modified

print("\nReturned values:")
print("New num returned:", result_num)
print("Modified list returned:", result_list)
```

```python
# 7. Write a program that performs basic arithmetic operations on two user-input numbers.

# Taking input from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Performing operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

# Handling division carefully to avoid division by zero
if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (cannot divide by zero)"

# Displaying the results
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

```python
# 8. Write a program to demonstrate the use of logical operators.

a = 10
b = 20
c = 5

# AND operator: True only if both conditions are True
print("AND Operator Examples:")
print(a < b and b > c)   # True and True → True
print(a > b and b > c)   # False and True → False

# OR operator: True if at least one condition is True
print("\nOR Operator Examples:")
print(a < b or a > b)    # True or False → True
print(a > b or b < c)    # False or False → False

# NOT operator: reverses the boolean value
print("\nNOT Operator Examples:")
print(not(a < b))        # not(True) → False
print(not(a > b))        # not(False) → True
```

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types.

# Taking input from the user (default type of input is string)
user_input = input("Enter something: ")

# Converting to integer
# This will work only if the user enters a valid number like 10, 25 etc.
try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to integer.")

# Converting to float
# This will work if input is a number like 3.14, 10, etc.
try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to float.")

# Converting to boolean
# In Python, any non-empty string becomes True, empty string becomes False
bool_value = bool(user_input)
print("Boolean value:", bool_value)
```

```python
# 10. Write code to demonstrate type casting with list elements
# Original list containing string numbers
numbers = ["10", "20", "30", "40"]

# Convert each element to integer using list comprehension
int_numbers = [int(i) for i in numbers]

# Convert each element to float
float_numbers = [float(i) for i in numbers]

# Convert each element to boolean
# Note: Non-empty strings always become True
bool_numbers = [bool(i) for i in numbers]

print("Original list:", numbers)
print("After casting to int:", int_numbers)
print("After casting to float:", float_numbers)
print("After casting to bool:", bool_numbers)
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.

total = 0  # variable to store the sum

for num in range(1, 51):  # loop from 1 to 50
    if num % 2 == 0:      # check if the number is even
        total += num      # add even number to total

print("Sum of even numbers from 1 to 50 is:", total)
```

```python
# 14. Write a program to reverse a string using a while loop.

# Take input from the user
text = input("Enter a string: ")

# Initialize an empty string for the reversed result
reversed_text = ""

# Set index to the last character of the string
i = len(text) - 1

# Loop backwards through the string
while i >= 0:
    reversed_text += text[i]   # Add characters in reverse order
    i -= 1                      # Move to the previous character

# Print the reversed string
print("Reversed string:", reversed_text)
```

```python
# 15.  Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))   # taking user input
factorial = 1                          # initial factorial value
i = 1                                  # counter variable

# factorial logic using while loop
while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)
```
